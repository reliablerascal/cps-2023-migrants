# 1. Read Data

```python
import pandas as pd
```

```python
df = pd.read_csv("./data/cps_change_sy2122_sy2223.csv",index_col=[0])
df.head()
```

# 2. Summary by Community Area

```python
df_community = df.groupby('community_area', as_index=False)\
    .agg({'d_total': sum,'d_low_income': sum,'d_hispanic': sum,'d_bilingual':sum})
```

```python
df_community.sort_values(by='d_total',ascending=False)[['community_area','d_total']]
```

```python
df_community.sort_values(by='d_low_income',ascending=False)[['community_area','d_low_income']]
```

```python
df_community.sort_values(by='d_hispanic',ascending=False)[['community_area','d_hispanic']]
```

```python
df_community.sort_values(by='d_bilingual',ascending=False)[['community_area','d_bilingual']]
```
